## バージョン管理システム:

中央集権型バージョン管理システム:
中央のリポジトリが全てのファイルの変更履歴を管理し、チームメンバーは中央リポジトリからファイルをチェックアウトして編集し、変更をコミットする。代表的な例としては、Subversion（SVN）があります。

分散型バージョン管理システム:
こちらのタイプのVCSでは、各開発者がローカルリポジトリを持ち、変更履歴を管理します。全てのローカルリポジトリは互いに連携し、変更を共有することができます。代表的な例としては、GitやMercurialがあります。Gitは特に広く使われており、分散型VCSの代表的な選択肢となっています。

バージョン管理システムには分散型を用いるべき理由

## 基本用語

### リポジトリ

リポジトリ：データや情報が集約、管理、保存される場所のこと


ローカルリポジトリ：開発者が自分のPC上にもつリポジトリ

リモートリポジトリ：ネットワーク上にある複数の開発者が共有するリポジトリ

### ブランチ

履歴の流れを分岐して記録してういくためのもの。  
分岐したブランチは他のブランチの影響を受けないため同じリポジトリ中で複数の変更を同時に進めることが可能




参考：https://backlog.com/ja/git-tutorial/stepup/01/

1. チームのメンバーは、他のメンバーの作業の影響を受けないように、メインのブランチから自分の作業専用のブランチを作成する  
2. 作業の終わったメンバーは、メインのブランチに自分のブランチの変更を取り込んでいく

このようにすることで、他のメンバーの作業による影響を受けることなく、自分の作業に取り込むこめる。  
また、作業単位で履歴を残すことで、問題が発生した場合に原因となる変更箇所の調査や対策を行うことが容易になる。

### 統合ブランチ・トピックブランチ

  
参考：https://backlog.com/ja/git-tutorial/stepup/02/  

統合ブランチ

・リリース版が何時でも作成可能なようしておくためのブランチ  
・トピックブランチの分岐元としても使用する  
・何らかの変更を行う場合は、トピックブランチを作成して作業を行う

トピックブランチ

・機能追加やバグ修正といったある課題に関する作業を行うために作成するブランチ  
・複数の課題に関する作業を同時に行う時は、その数だけトピックブランチが作成される

1. add // セーブしたい内容を追加
2. add . //変更した全てのファイルを追加する
3. commit // セーブポイント commit -m "コメント" // コメント
4. push // アップロード
